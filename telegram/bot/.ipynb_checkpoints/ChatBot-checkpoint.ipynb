```python
import sys
# import ChatBotModel
import telepot
from telepot.loop import MessageLoop
from telepot.namedtuple import InlineKeyboardMarkup as MU # 마크업
from telepot.namedtuple import InlineKeyboardButton as BT # 버튼
from telegram.ext import Updater, MessageHandler, Filters, CommandHandler

token = '5035520069:AAHXrviXBFO9didF_6HEpE1WhBeLuw-rTUM'
chat_id = '@hermes_test'
bot = telepot.Bot(token)

updater = Updater(token=token, use_context=True) 
dispatcher = updater.dispatcher 

def check_id(bot, update):
    try:
        id = update.message.chat.id
        print('Chat ID', id)
        return id
    except:
        id = update.channel_post.chat.id
        return id

def check_nickname(bot, update):
    try:
        nickname = update.message.from_user.first_name
        print('Chat NickName', nickname)
        return nickname
    except:
        nickname = update.channel_post.from_user.first_name
        return nickname

def start_command(bot, update):
    id = check_id(bot, update)
    nickname = check_nickname(bot, update)
    bot.send_message(chat_id=id, text="Hello " + nickname +"! Welcome!\n\n")

def btn_show(msg):
    if msg['text'] == "link":
        btn1 = BT(text = "Website", url = "https://xd.adobe.com/view/2314682e-3f11-491b-9c1e-1fe804a3e242-7647/", callback_data = "1")
        btn2 = BT(text = "Twitter", url = "https://twitter.com/official_LUCK_", callback_data = "2")
        btn3 = BT(text = "Reddit", url = "https://www.reddit.com/user/official_LUCK_", callback_data = "3")
        mu = MU(inline_keyboard = [[btn1], [btn2], [btn3]])
        bot.sendMessage(chat_id, "Choose one", reply_markup = mu)

def query_ans(msg):
    query_id = msg["id"]
    query_data = msg["data"]
    if query_data == "1":
        bot.answerCallbackQuery(query_id, text = "Website")
    elif query_data == "2":
        bot.answerCallbackQuery(query_id, text = "Twitter")
    elif query_data == "3":
        bot.answerCallbackQuery(query_id, text = "Reddit")

def start(update, context): 
    context.bot.send_message(chat_id=update.effective_chat.id, text="Get $LUCK & Getting Lucky!") 
    start_handler = CommandHandler('start', start) 
    dispatcher.add_handler(start_handler)

print('start telegram chat bot')


MessageLoop(bot, {'chat' : btn_show, 'callback_query' : query_ans}).run_as_thread()
updater.start_polling()
updater.idle()


# def greeting(bot, update):
#     hermes.sendMessage('Welcome,')
#     sound = naming()
#     hermes.sendMessage(sound)
#     hermes.sendMessage('!!')

# def goodbye(bot, update):
#     hermes.sendMessage('Good Bye!')
# #     hermes.stop()

# def naming():
#     return 'Loop!'



# hermes = ChatBotModel.BotHermes()
# hermes.add_handler('link', btn_show)
# hermes.add_handler('greeting', greeting)
# hermes.add_handler('stop', goodbye)
# hermes.start()
```

```python
from telegram.ext import Updater, MessageHandler, Filters, CommandHandler  # import modules

my_token = '5035520069:AAHXrviXBFO9didF_6HEpE1WhBeLuw-rTUM'

print('start telegram chat bot')


# message reply function
def get_message(update, context) :
    update.message.reply_text("got text")
    update.message.reply_text(update.message.text)


# help reply function
def help_command(update, context) :
    update.message.reply_text("무엇을 도와드릴까요?")


updater = Updater(my_token, use_context=True)

message_handler = MessageHandler(Filters.text & (~Filters.command), get_message) # 메세지중에서 command 제외
updater.dispatcher.add_handler(message_handler)

help_handler = CommandHandler('help', help_command)
updater.dispatcher.add_handler(help_handler)

updater.start_polling(timeout=3, clean=True)
updater.idle()
```
